```
# Install plotly in colab
!pip install plotly

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load your cleaned dataset
df = pd.read_csv("Amazon Sale Report.csv")

# Display first few rows
df.head()
```


### 1. Dataset Overview
Before diving into visualizations, let's get a general understanding of the dataset
— including the number of rows, columns, and basic statistics.


## **2**. Univariate Analysis

```
plt.figure(figsize=(6,4))
sns.countplot(x='Status', data=df, palette='viridis')
plt.title("Order Status Distribution")
plt.xticks(rotation=45)
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.countplot(x='Fulfilment', data=df, palette='Set2')
plt.title("Fulfilment Method Distribution")
plt.xticks(rotation=45)
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.countplot(x='Sales Channel', data=df, palette='pastel')
plt.title("Sales Channel Distribution")
plt.show()
```

```
plt.figure(figsize=(8,5))
df['Category'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Categories by Count")
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.histplot(df['Qty'], bins=30, kde=True, color='orange')
plt.title("Quantity Ordered Distribution")
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.histplot(df['Amount'], bins=30, kde=True, color='green')
plt.title("Order Amount Distribution")
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.scatterplot(x='Qty', y='Amount', data=df, alpha=0.6)
plt.title("Amount vs Quantity")
plt.show()
```

```
plt.figure(figsize=(6,4))
df.groupby('Fulfilment')['Amount'].mean().sort_values().plot(kind='bar', color='purple')
plt.title("Average Amount by Fulfilment Method")
plt.show()
```

```
plt.figure(figsize=(6,4))
df.groupby('Sales Channel')['Amount'].mean().sort_values().plot(kind='bar', color='teal')
plt.title("Average Amount by Sales Channel")
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.heatmap(df[['Qty', 'Amount']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Amount'].sum()
monthly_sales.plot(kind='line', marker='o', figsize=(8,5), color='blue')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.show()
```

```
top_rev = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(10)
top_rev.plot(kind='bar', figsize=(8,5), color='coral')
plt.title("Top 10 Categories by Revenue")
plt.show()
```

#  **Conclusion**
From the above analysis:
- The most common statuses, fulfilment methods, and sales channels were identified.
- Amount and quantity show some correlation.
- Certain categories and months dominate sales revenue.
- Seasonal trends are visible in the monthly sales line chart.
