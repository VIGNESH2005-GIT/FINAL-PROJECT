# **Step 6 : Handling the missing values**

```
# 📌 Step 6: Dropping Columns with Too Many Missing Values
columns_to_drop = ['Unnamed: 22', 'promotion-ids', 'fulfilled-by']
columns_to_drop_exist = [col for col in columns_to_drop if col in df.columns]
df.drop(columns_to_drop_exist, axis=1, inplace=True)
```

```
df.info()
```

```
df.isnull().sum()
```

```
# Handling Remaining Missing Values
df['Courier Status'] = df['Courier Status'].fillna('Unknown')

# Forward fill remaining nulls
df.ffill(inplace=True)
```

```
#  Remove Duplicate Rows (if any)
df.drop_duplicates(inplace=True)
```

```
# Reset Index After Cleaning
df.reset_index(drop=True, inplace=True)
```
